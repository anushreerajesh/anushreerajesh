```python
from google.colab import files

uploaded = files.upload()

image_path = "/content/face_sample1.jpg"
import os

if (os.path.exists(image_path)):
    print("✅ File found!")
else:
    print("❌ File not found. Upload again.")
import cv2
import numpy as np
from mtcnn import MTCNN
import matplotlib.pyplot as plt

def detect_faces(image_path):
    # Load image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB

    # Initialize MTCNN detector
    detector = MTCNN()
    faces = detector.detect_faces(image)

    for face in faces:
        x, y, width, height = face['box']
        keypoints = face['keypoints']

        # Draw bounding box
        cv2.rectangle(image, (x, y), (x+width, y+height), (0, 255, 0), 2)

        # Draw keypoints
        for key, point in keypoints.items():
            cv2.circle(image, point, 2, (255, 0, 0), -1)

    return image, faces

# Test the function
image_path = 'face_sample1.jpg'  # Replace with your image path
processed_image, detected_faces = detect_faces(image_path)

# Display the image with detections
plt.figure(figsize=(8, 6))
plt.imshow(processed_image)
plt.axis('off')
plt.show()

# Print detected face details
print("Detected Faces:", detected_faces)
```
